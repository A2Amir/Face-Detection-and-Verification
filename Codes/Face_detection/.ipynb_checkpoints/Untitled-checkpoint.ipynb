```python
default_color= (0,255,0)
default_thickness = 3
```

```python
min_size = 20
thresholds = [.6, .7, .7]
pyramid_factor = .709
```

```python
def draw_boxes(img, boxes):
    for bounding_box in boxes:
        pts = bounding_box[:4].astype(np.int32)
        pts1 = (pts[0],pts[1])
        pts2 = (pts[2],pts[3])
        drawn_img = np.copy(img)
        cv2.rectangle(drawn_img, pts1, pts2, default_color, default_thickness)
    return drawn_img
```

```python
with tf.Graph().as_default():
    sess = tf.compat.v1.Session()
    pnet,rnet,onet = detect_face.create_mtcnn(sess, None)
    
```

```python
img = plt.imread('C:/Users/ziaeeamir/Desktop/Face/Face-Recognition-and-Verification/imgs/5.jpg')
plt.imshow(img)
```

```python
boxes, points = detect_face.detect_face(img, min_size, pnet, rnet, onet, thresholds, pyramid_factor)
```

```python
idetected_img = draw_boxes(img, boxes)
plt.imshow(idetected_img)
```
