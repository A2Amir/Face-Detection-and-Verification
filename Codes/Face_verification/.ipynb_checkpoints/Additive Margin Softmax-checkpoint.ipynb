# Biuld Additive Margin Softmax Layer

```python
class AMSoftmax(Layer):
    def __init__(self, units, **kwargs):
        self.units = units
        self.kernel = None
        super(AMSoftmax,self).__init__(**kwargs)

    def build(self, input_shape):
        assert len(input_shape) >= 2

        self.kernel = self.add_weight(name='kernel',
                                      shape=(input_shape[1], self.units),
                                      initializer='uniform',
                                      trainable=True)
        self.kernel = K.l2_normalize(self.kernel, axis=0)
        super(AMSoftmax, self).build(input_shape)

    def call(self, inputs, **kwargs):
        # get cosine similarity
        # cosine = x * w / (||x|| * ||w||)
        inputs = K.l2_normalize(inputs, axis=-1)
        kernel =self.kernel  # K.l2_normalize(self.kernel, axis=0)
        cosine = K.dot(inputs, kernel)
        return cosine

    def compute_output_shape(self, input_shape):
        return input_shape[0], self.units

    def get_config(self):
        config = {
            'units': self.units}
        base_config = super(AMSoftmax, self).get_config()

        return dict(list(base_config.items())
                    + list(config.items()))


# reference:
# https://github.com/hao-qiang/AM-Softmax/blob/master/AM-Softmax.ipynb
def amsoftmax_loss_v1(y_true, y_pred, scale=30.0, margin=0.35):
    # make two constant tensors.
    m = K.constant(margin, name='m')
    s = K.constant(scale, name='s')
    # reshape the label
    label = K.reshape(K.argmax(y_true, axis=-1), shape=(-1, 1))
    label = K.cast(label, dtype=tf.int32)

    pred_batch = K.reshape(tf.range(K.shape(y_pred)[0]), shape=(-1, 1))
    # concat the two column vectors, one is the pred_batch, the other is label.
    ground_truth_indices = tf.concat([pred_batch,
                                      K.reshape(label, shape=(-1, 1))], axis=1)
    # get ground truth scores by indices
    ground_truth_scores = tf.gather_nd(y_pred, ground_truth_indices)

    # if ground_truth_score > m, group_truth_score = group_truth_score - m
    added_margin = K.cast(K.greater(ground_truth_scores, m),
                          dtype=tf.float32) * m
    added_margin = K.reshape(added_margin, shape=(-1, 1))
    added_embedding_feature = tf.subtract(y_pred, (tf.cast(y_true,dtype=tf.float32))  * added_margin) * s

    cross_entropy = tf.nn.softmax_cross_entropy_with_logits(labels=y_true,
                                                               logits=added_embedding_feature)
    loss = tf.reduce_mean(cross_entropy)
    return loss

def amsoftmax_loss_v2(y_true, y_pred, scale=30.0, margin=0.35):
    # make two constant tensors.
    margin = K.constant(margin, name='m', dtype=tf.float32)
    scale = K.constant(scale, name='s', dtype=tf.float32)
    #print(y_true.dtype, y_pred.dtype,margin.dtype,scale.dtype)
    y_pred = tf.cast(y_true,dtype=tf.float32) * (y_pred - margin) + (1 - (tf.cast(y_true,dtype=tf.float32))) * y_pred
    y_pred *=  scale
    return K.categorical_crossentropy(y_true, y_pred, from_logits=True)
```

# Build Model

```python
base_model = VGG16(weights='imagenet',  # Load weights pre-trained on ImageNet.
                    input_shape=(160, 160, 3),
                    include_top=False)  # Do not include the ImageNet classifier at the top.
base_model.trainable = False

inputs = Input(shape=(160, 160, 3))
                  
# Getting output tensor of the last VGG layer that we want to include
x = base_model(inputs, training=False)
x = Flatten() (x)


x = Dense(1024, activation=None, name='dense1')(x)
x = PReLU() (x)
x = BatchNormalization() (x)

x = Dropout(0.5)(x)
x = Dense(512, activation=None, name='dense2')(x)
x = PReLU() (x)
x = BatchNormalization() (x)

x = Dense(256, name='dense3')(x)
output = AMSoftmax(105)(x)

model = Model(inputs, output)
```

```python
# Do not forget to compile it
model.compile(loss=amsoftmax_loss_v1,
                     optimizer='adam',
                     metrics=['accuracy'])
```

# Test Model on the Fashion MNIST dataset

```python
def euclidean_distance(vects):
    x,y = vects
    sum_square = K.sum(K.square(x-y), axis=1, keepdims=True)
    sqrt = K.sqrt(K.maximum(sum_square, K.epsilon()))
    return sqrt
```

```python
num_classes = max(y_test) + 1
digit_indices = [np.where(y_test == i)[0] for i in range(num_classes)]

f_idx = random.randint(0, num_classes-1)
pair1 = x_test[random.choice(digit_indices[f_idx])]
pair2 = x_test[random.choice(digit_indices[f_idx])]
pair3 = x_test[random.choice(digit_indices[f_idx+2])]
```

```python
# plot pairs
fig = plt.figure(figsize=(25,25))

plt.subplot(1, 3, 1)
plt.imshow(pair1)

plt.subplot(1, 3, 2)
plt.imshow(pair2)

plt.subplot(1, 3, 3)
plt.imshow(pair3)


print(pair1.shape)
```

```python
def compare(image1, image2, threshold = 1.18):
    input1 = np.expand_dims(image1,[0])
    input2 = np.expand_dims(image2,[0])
    
    #make prediction
    vec1 = func([input1])
    vec2 = func([input2])
    #check distance
    distance = euclidean_distance([vec1[0], vec2[0]])
    is_same = distance.numpy()[0] < threshold
    print(is_same, 'distance:',distance.numpy()[0])
    print()
```

```python
compare(pair1, pair2)
```
